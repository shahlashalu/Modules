```python
# What is Module in Python?
# The module is a simple python file that contains collections of functions and global variables and with having a .py extension file.
# It is an executable file and to organize all the modules we have the concept called Package in Python

# Examples of modules:
# Datetime
# Regex
# Random
# calendar
```

```python

# What is Package in Python?

# The Package is a simple directory having collections of modules. This directory contains simple modules and.
# The Package is simply a namespace. The package also contains sub-packages inside it.

# Examples of Packages
# Numpy
# Pandas

```

```python
# What is Library in Python?
# The Library is having a collection of related functionality of cides that allows you to perform many tasks without writung your code.
# It is a reusable chunk of code that we can use by importing it into our program, we can just use it by importing that library and
# calling the method of that library with a period(.). However it is often assumed that while a package is a collection of modules,
# a library is a collection of packages.

# Examples of libraries:
# Matplotlib
# Pytorch
# Pygame
# Seaborn
```

```python
#password

import re
pwd = input('enter the password : ')

if len(pwd)<6 or len(pwd)>16:
  print('your password is not satisfied.')

elif not re.search('[a-z]',pwd):
  print('no small letters are found')

elif not re.search('[A-Z]',pwd):
  print('no capital letters are found')

elif not re.search('[0-9]',pwd):
  print('no digits are available')

elif not re.search('[^a-z,A-Z,0-9,\s]',pwd):
  print('no spl characters are found')

else:
  print('successfully completed')


print(pwd)
```
